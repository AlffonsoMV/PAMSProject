```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 6.67430e-11

class CelestialBody:
    def __init__(self, name, mass, pos, momentum):
        self.name = name
        self.mass = mass
        self.pos = np.array(pos, dtype=float)
        self.momentum = np.array(momentum, dtype=float)

    def compute_force(self, other_body):
        r = np.linalg.norm(self.pos - other_body.pos)
        f = -(G * self.mass * other_body.mass / r**3) * (self.pos - other_body.pos)
        return f

    def update_position(self, dt):
        self.pos += dt * self.momentum / self.mass

    def update_momentum(self, dt, force):
        self.momentum += dt * force

def stormer_verlet_step(bodies, dt):
    # Half-step update for momentum
    for body in bodies:
        total_force = np.zeros(2)
        for other_body in bodies: 
            if body != other_body:
                total_force += body.compute_force(other_body)
        body.update_momentum(0.5 * dt, total_force)

    # Full-step update for positions
    for body in bodies:
        body.update_position(dt)

    # Another half-step update for momentum
    for body in bodies:
        total_force = np.zeros(2)
        for other_body in bodies:
            if body != other_body:
                total_force += body.compute_force(other_body)
        body.update_momentum(0.5 * dt, total_force)

def simulate(bodies, T, dt):
    times = np.arange(0, T, dt)
    positions = {body.name: [] for body in bodies}

    for t in times:
        for body in bodies:
            positions[body.name].append(body.pos.copy())

        stormer_verlet_step(bodies, dt)

    return positions

def plot_simulation(positions):
    plt.figure(figsize=(10, 10))
    for name, pos in positions.items():
        pos = np.array(pos)
        plt.plot(pos[:, 0], pos[:, 1], label=name)
    plt.legend()
    plt.grid(True)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Solar System Simulation")
    plt.show()


# Data from https://nssdc.gsfc.nasa.gov/planetary/factsheet/
bodies = [
    CelestialBody("Sun", 1.989e30, [0, 0], [0, 0]),
    CelestialBody("Mercury", 3.301e23, [0, 5.7e10], [47.4e3 * 0.33e24, 0]),
    CelestialBody("Venus", 4.867e24, [0, 1.08e11], [35.0e3 * 4.87e24, 0]),
    CelestialBody("Earth", 5.972e24, [0, 1.49e11], [29.8e3 * 5.97e24, 0]),
    CelestialBody("Mars", 6.417e23, [0, 2.28e11], [24.1e3 * 0.642e24, 0]),
    #CelestialBody("Jupiter", 1.899e27, [0, 7.78e11], [1.31e4 * 1.899e27, 0]),
    #CelestialBody("Saturn", 5.685e26, [0, 1.43e12], [9.69e3 * 5.685e26, 0]),
    #CelestialBody("Uranus", 8.682e25, [0, 2.87e12], [6.81e3 * 8.682e25, 0]),
    #CelestialBody("Neptune", 1.024e26, [0, 4.50e12], [5.43e3 * 1.024e26, 0]),
]

positions = simulate(bodies, 365 * 24 * 3600, 24 * 3600)  # 1 year with 1-day step
plot_simulation(positions)
```

```python
def gravitational_potential_at_point(point, body):
    r = np.linalg.norm(point - body.pos)
    if r == 0:  
        return 0
    return -G * body.mass / r

def total_potential_at_point(point, bodies):
    total_potential = 0
    for body in bodies:
        total_potential += gravitational_potential_at_point(point, body)
    return total_potential

def potential_map(bodies, x_range, y_range, resolution=100):
    x = np.linspace(x_range[0], x_range[1], resolution)
    y = np.linspace(y_range[0], y_range[1], resolution)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X)

    for i in range(resolution):
        for j in range(resolution):
            Z[i, j] = total_potential_at_point(np.array([X[i, j], Y[i, j]]), bodies)

    return X, Y, Z

def plot_potential_map(bodies):
    X, Y, Z = potential_map(bodies, [-3e11, 3e11], [-3e11, 3e11])
    plt.figure(figsize=(10, 10))
    plt.contourf(X, Y, Z, 100, cmap='inferno')
    plt.colorbar().set_label('Potential (Joules)')
    for body in bodies:
        plt.scatter(body.pos[0], body.pos[1], label=body.name)
    plt.legend()
    plt.title('Gravitational Potential Map')
    plt.xlabel('x (m)')
    plt.ylabel('y (m)')
    plt.show()

plot_potential_map(bodies)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def potential_map_3d(bodies, x_range, y_range, resolution=100):
    x = np.linspace(x_range[0], x_range[1], resolution)
    y = np.linspace(y_range[0], y_range[1], resolution)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X)

    for i in range(resolution):
        for j in range(resolution):
            Z[i, j] = total_potential_at_point(np.array([X[i, j], Y[i, j]]), bodies)

    return X, Y, Z

def plot_potential_map_3d(bodies):
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')

    X, Y, Z = potential_map_3d(bodies, [-3e11, 3e11], [-3e11, 3e11])
    ax.plot_surface(X, Y, Z, cmap='inferno', edgecolor='none')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Potential (Joules)')
    ax.set_title('3D Gravitational Potential Map')
    
    for body in bodies:
        ax.scatter(body.pos[0], body.pos[1], 0, label=body.name, s=50)

    ax.legend()
    plt.show()

plot_potential_map_3d(bodies)
```

```python
def total_energy(bodies):
    kinetic = sum(0.5 * np.transpose(body.momentum).dot((np.identity(2) * 1 / body.mass).dot(body.momentum)) for body in bodies)
    potential = sum(body.mass * total_potential_at_point(body.pos, bodies) for body in bodies)
    return kinetic + potential

def plot_energy_over_time(bodies, T, dt):
    times = np.arange(0, T, dt)
    energies = []

    for t in times:
        energies.append(total_energy(bodies))
        stormer_verlet_step(bodies, dt)

    plt.plot(times, energies)
    plt.xlabel("Time (s)")
    plt.ylabel("Total Energy (J)")
    plt.title("Energy of the Solar System over Time")
    plt.grid(True)
    plt.show()

plot_energy_over_time(bodies, 10 * 365 * 24 * 3600, 24 * 3600)
```
